```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

class k_means():

    col = {1: 'lime' , 2: 'black', 3: 'blue', 4: 'red', 5: 'm', 6: 'c', 7: 'brown', 8:'deeppink', 9: 'g'} # colours for the different clusters
    shapes = {1: 'X',2: '*',3: 'h',4: 'P', 5: 'D' , 6: 'd', 7: 's', 8: 'h', 9: '^'} # symbols for the centroids
    size = 90 # size for the centroids


    def __init__(self, filename, k, MaxIters):
        self.df = pd.read_csv(filename, header=None) # data is loaded in a dataframe
        self.k = k # number of clusters are stored
        self.MaxIters= MaxIters # number for Max iterations are stored
        self.run() # call run() function

        #calculate distance and assign the 4 closest points to the clustered centers
    def assign(self,df,centroids):
        for i in centroids.keys():
            #sqqrt((x1-x2)^2 - (y1-y1)^2)
            df['distance_from_{}'.format(i)] = (np.sqrt((df[0] - centroids[i][0]) **2 
            + (df[1] - centroids[i][1]) **2)) # calculates the distance from all points
        centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()] # stores every distance from each centroid
        df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1) # takes the minimum distance
        df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_'))) # takes the (i), the centroid number it is closest to
        df['color'] = df['closest'].map(lambda x: self.col[x]) # colour is taken from list
        return df                
        
    def update(self, centroids):
        for i in self.centroids.keys():
            self.centroids[i][0] = np.mean(self.df[self.df['closest'] == i][0]) # mean of the closest x points
            self.centroids[i][1] = np.mean(self.df[self.df['closest'] == i][1]) # mean of the closest y points
        return centroids

    def run(self):   
        self.centroids = {
            i+1: [np.random.randint(0,4000000), np.random.randint(0,4000000)] # generate random points
            for i in range(self.k)
        }
        self.df = self.assign(self.df, self.centroids) # call assign() function
        self.centroids = self.update(self.centroids) # call update() function

        iters = 0
        while True and iters <= self.MaxIters:
            closest_centroids = self.df['closest'].copy(deep=True) # copy previous centroids
            self.centroids = self.update(self.centroids) # call update()
            self.df = self.assign(self.df, self.centroids) # call assign()
            if closest_centroids.equals(self.df['closest']): # previous centroid = current centroid
                self.plots()
                break
        iters+=1

    def plots(self):

        fig = plt.figure(figsize = (14,14)) # size of the diagram
        plt.scatter(self.df[0], self.df[1], s = 50, alpha = 0.5, color = self.df['color'], edgecolor = 'k') #scatter points
        for i in self.centroids.keys():
            plt.scatter(*self.centroids[i], s = self.size, color = self.col[i], marker = self.shapes[i]) # scatter centroids
        plt.xlim(0,4000000) # x axis size
        plt.ylim(0,4000000) # y axis size
        plt.show()



                
```

Pandas DataFrame is easier to manipulate and analyze with this particular amount of data, therefore I decided to use pandas. 

Col, shapes and size is initialised. col has a list with colours assigned to numbers 1 through 9 that will be used for the different clusters. The shapes has a list of symbols that is assigned to numbers 1 through 9 that will be used to differentiate the centroids. Size will be used to make the centroids larger than the other points. 

class k_means only accepts 3 arguments; the filename,number of clusters and the maximum iteration. It calls on the run() function, where k random points are generated for the centroids and stores them as a list. 
The function assign() is called to calculate the distance from the points to every centroid and stores them with the centroid number it is closest to with the colour assigned to that number. The function update() is called to calculate the mean of the closest points.

Iters is then initialised for the while loop. The centroids will continuously be updating in this loop with the update() and assign() functions until the loop ends. The while loop will end when Iters equals 300 or when the previous centroid equals the present centroid and it will call the function plots() before the while loop ends. The data is then plotted on the diagram.

The elbow method is used to determine the maximum amount of clusters that data can be clustered in to. This method uses the Euclidean distance metric to calculate the mean of the squared distances from the center of the clusters of the respective clusters.
From the visual representation of the results, the number of clusters are selected at the "elbow". This is the point where the decrease of inertia starts to become linear.

EG: k(number of cluster) = 4


```python
import pandas as pd
df = pd.read_csv("data.txt", header = None)
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import cluster

numClusters = range(1,10)
SSE = []
for k in numClusters:
    k_means = cluster.KMeans(n_clusters=k)
    k_means.fit(df)
    SSE.append(k_means.inertia_)

plt.plot(numClusters, SSE)
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
```
