```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

class kmeans():

    df = pd.read_csv("data.txt", header = None) #dataframe for reading in data
    k =  0 #num of clusters
    MaxIters = 0 #limit of iterations
    
    #Output
    colmap = {1: 'r' , 2: 'g', 3: 'b', 4: 'y'}
    m = {1: 'X',2: '*',3: 'h',4: 'P'}
    s = [90]
    
    #initialise centroids
    centroids = {}

    def __init__(self, filename, k, MaxIters):
        self.df = pd.read_csv(filename, header = None)#data frame
        self.k = k
        self.MaxIters = MaxIters
        self.run()
       
        
    #calculate distance and assign the 4 closest points to the clustered centers
    def assign(self,df,centroids):
        for i in centroids.keys():
            #sqqrt((x1-x2)^2 - (y1-y1)^2)
            df[f'distance_from_{i}'] = (np.sqrt(((df[0] - centroids[i][0])**2) 
            + ((df[1] - centroids[i][1])**2)))
            
        centroid_distance_cols = [f'distance_from_{i}' for i in centroids.keys()]
        df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)
        df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
        df['color'] = df['closest'].map(lambda x: self.colmap[x])
        return df


    def update(self, centroids):
        for i in self.centroids.keys():
            self.centroids[i][0] = np.mean(self.df[self.df['closest'] == i][0])
            self.centroids[i][1] = np.mean(self.df[self.df['closest'] == i][1])
        return centroids

    def run(self):   
        self.centroids = {
            i+1: [np.random.randint(0,4000000), np.random.randint(0,4000000)]
            for i in range(self.k)
        }
        self.df = self.assign(self.df, self.centroids)
        self.centroids = self.update(self.centroids)

        iters = 0
        change = False
        
        while (change != True) and (iters <= self.MaxIters):
            closest_centroids = self.df['closest'].copy(deep = True)
            self.centroids = self.update(self.centroids)
            self.df = self.assign(self.df, self.centroids)
            if closest_centroids.equals(self.df['closest']):
                change = True
                break
                
        iters += 1
        
    def printPlot(self):
        fig = plt.figure(figsize = (12,12))
        plt.scatter(self.df[0],self.df[1], color = self.df['color'], alpha = 0.6, edgecolor = 'k')
        for i in self.centroids.keys():
            plt.scatter(*self.centroids[i],color = self.colmap[i],marker = self.m[i], s = self.s)
        plt.xlim(0,4000000)
        plt.ylim(0,4000000)
        plt.show()

                
```

Task 3

Pandas DataFrame is easier to manipulate and analyze this particular amount of data. Therefore I decided to load the data using pandas. class kmeans only accepts 3 arguments; the filename,number of clusters and the maximum iteration. It will then select k random points and set it as centroids and call on functions assign() and update(). It will calculate the closest points to the cluster centers and update the centroids. To repeat this until there are no more updates that can be done, a while loop is implemented so that it calls the functions assign() and update(). This will continue to run until the centroid is the center of the cluster or until iters = MaxIters. There is a a function called printPlot() that plots the data with the clusters being a different colour according to the centroid and the centroids are differentiated with different colour and shape.

Task 4

The elbow method is used to determine the maximum amount of clusters that data can be clustered in to. This method uses the Euclidean distance metric to calculate the mean of the squared distances from the center of the clusters of the respective clusters.
From the visual representation of the results, the number of clusters are selected at the "elbow". This is the point where the decrease of inertia starts to become linear.
EG: k(number of cluster) = 4

```python
import pandas as pd
df = pd.read_csv("data.txt", header = None)
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import cluster

numClusters = range(1,10)
SSE = []
for k in numClusters:
    k_means = cluster.KMeans(n_clusters=k)
    k_means.fit(df)
    SSE.append(k_means.inertia_)

plt.plot(numClusters, SSE)
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
```
