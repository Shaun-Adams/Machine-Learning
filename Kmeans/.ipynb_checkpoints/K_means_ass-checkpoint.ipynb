## Task 1

```python
import copy
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

class kmeans():



    #global variables
    #----------------------------------------------------------------------------
    k = 0#number of centroids
    df=pd.read_csv("data.txt", header=None)#data frame

    # centroids
    centroids={}

    colmap = {1: 'r', 2: 'g', 3: 'b', 4: 'c', 5: 'm',  6:'y', 7:'k'}

    #functions
    #----------------------------------------------------------------------------
    def __init__(self, filename, k):
        self.df = pd.read_csv(filename, header=None)#data frame
        self.k = k
        self.run()
       
    
    
    def assignment(self,df, centroids):#computes and assigns the distance to each centroid to a dataframe for each (x,y)
        for i in centroids.keys():
            # sqrt((x1 - x2)^2 - (y1 - y2)^2)
            df['distance_from_{}'.format(i)] = (
                np.sqrt(
                    (df[0] - centroids[i][0]) ** 2 + (df[1] - centroids[i][1]) ** 2
                )#computes distance from centroid and adds to distance column
            )
        centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
        df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)#gets closest centroid number and adds it to closest centroid column in dataframe
        df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
        df['color'] = df['closest'].map(lambda x: self.colmap[x])
        return df


    def update(self,centroids):#updates centroid values
        for i in self.centroids.keys():
            self.centroids[i][0] = np.mean(self.df[self.df['closest'] == i][0])#updates x of centroid
            self.centroids[i][1] = np.mean(self.df[self.df['closest'] == i][1])#updates y of centroid
        return centroids

    def run(self):
        self.centroids = {
            i+1: [np.random.randint(0,4000000), np.random.randint(0, 4000000)]
            for i in range(self.k)
         }
        
        self.df = self.assignment(self.df, self.centroids)
        self.centroids = self.update(self.centroids)

        # Continue until all assigned categories don't change any more
        while True:# or for i in range(number of iterations)
            closest_centroids = self.df['closest'].copy(deep=True)
            self.centroids = self.update(self.centroids)
            self.df = self.assignment(self.df, self.centroids)
            if closest_centroids.equals(self.df['closest']):
                break
        # or use instead of while loop
        #for i in range(200):# for 200 iterations
        #    closest_centroids = self.df['closest'].copy(deep=True)
        #    self.centroids = self.update(self.centroids)
        #    self.df = self.assignment(self.df, self.centroids)

    def printPlot(self):
        fig = plt.figure(figsize=(10, 10))
        plt.scatter(self.df[0], self.df[1], color=self.df['color'], alpha=0.5, edgecolor='k')
        for i in self.centroids.keys():
            plt.scatter(*self.centroids[i], color=self.colmap[i])
        plt.xlim(0, 4000000)
        plt.ylim(0, 4000000)
        plt.show()

#for task 3 to do elbow method... (sum of square error calculation is wron somehow.....)
    def getSSE(self):
        n=0
        sum1=0
        sum2=0
        centroid_distance_cols = ['distance_from_{}'.format(i) for i in self.centroids.keys()]
        #print(centroid_distance_cols)
        for col in centroid_distance_cols:
            for num in self.df[col]:
                n+=1
                sum1=sum1+num
                sum2=sum2+num**2
                #print(num)
        sse=((1/n)*sum1**2)-sum2
        return sse
                
    
```
